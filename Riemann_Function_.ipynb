```python
import numpy as np
import matplotlib.pyplot as plt

def riemann_function(x, N=100):
    f = np.zeros_like(x)
    for n in range(1, N + 1):
        f += np.sin(n**2 * x) / (n**2)
    return f

# Parameters for plotting
x_min = 0
x_max = 10  # Adjust interval as needed (e.g., [0, 2*np.pi] for periodicity-like view)
num_points = 10000  # Number of points for smooth plotting
N_terms = 100  # Number of terms; try 200 for more roughness (but slower)

# Generate x values
x = np.linspace(x_min, x_max, num_points)

# Compute f(x)
y = riemann_function(x, N=N_terms)

# Plot the graph
plt.figure(figsize=(12, 6))
plt.plot(x, y, 'b-', linewidth=1, label=f'Riemann function (N={N_terms} terms)')
plt.title('Riemann Function: $f(x) = \\sum_{n=1}^\\infty \\frac{\\sin(n^2 x)}{n^2}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.ylim(-0.5, 0.5)  # Adjust y-limits to focus on the function (it oscillates around 0)
plt.show()

# Optional: Print some values for verification
print(f"Sample values:")
print(f"f(0) ≈ {y[0]:.4f} (should be 0)")
print(f"f(π/2) ≈ {riemann_function(np.pi/2, N_terms):.4f}")
```
