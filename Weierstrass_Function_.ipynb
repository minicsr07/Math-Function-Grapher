```python
import numpy as np
import matplotlib.pyplot as plt

def weierstrass(x, a=0.5, b=5, N=20):
    total = 0
    for n in range(N):
        total += a*n * np.cos(b*n * np.pi * x)
    return total
x_values = np.arange(-2, 2.01, 0.01)

y_values = np.array([weierstrass(x) for x in x_values])

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, color='blue')
plt.title('Weierstrass Function')
plt.xlabel('x')
plt.ylabel('W(x)')
plt.grid(True)
plt.show()
```
